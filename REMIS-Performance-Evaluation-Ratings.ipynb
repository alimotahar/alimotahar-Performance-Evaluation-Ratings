```python
corr_matrix = ConvertREMIS.corr().abs()

threshold = 0.5

mask = (corr_matrix > threshold)

corr_pairs = []
for i in range(len(mask.columns)):
    for j in range(i):
        if mask.iloc[i, j]:
            corr_pairs.append((mask.columns[i], mask.columns[j], corr_matrix.iloc[i, j]))

corr_pairs_df = pd.DataFrame(corr_pairs, columns=['Feature 1', 'Feature 2', 'Correlation'])
corr_pairs_df = corr_pairs_df.sort_values(by='Correlation', ascending=False)

corr_pairs_df
```

# MODEL CREATION & ANALYSIS

* Define Dependant & Independant Variabels
* Split the Data to Train-Set and Test-Set
* Model creation, prediction and Analysis

## 1. Define Dependant & Independant Variabels

```python
# Define Independants (X) and Dependant (y) Variables
X = ConvertREMIS.drop('Performance_Evaluation_Score', axis=1)
y = ConvertREMIS['Performance_Evaluation_Score']
```

## 2. Split the Data to Train-Set and Test-Set

```python
from sklearn.model_selection import train_test_split

# Split the data into a 70% train set and a 30% test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## 3. Model creation, prediction and Analysis

* Linear Regression
* Decision Tree Regressor
* Random Forest Regressor
* XGBoost Regressor

* Comparison of Machine Learning Models' Outcome.

### 3.1 Linear Regression:

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

lr_model = LinearRegression()

lr_model.fit(X_train, y_train)

lr_predictions = lr_model.predict(X_test)

lr_mse = mean_squared_error(y_test, lr_predictions)
lr_mae = mean_absolute_error(y_test, lr_predictions)
lr_r2 = r2_score(y_test, lr_predictions)

print("Linear Regression Results:")
print("Mean Squared Error:", lr_mse)
print("Mean Absolute Error:", lr_mae)
print("R-squared:", lr_r2)
```

### 3.2 Decision Tree Regressor:

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor()

dt_model.fit(X_train, y_train)

dt_predictions = dt_model.predict(X_test)

dt_mse = mean_squared_error(y_test, dt_predictions)
dt_mae = mean_absolute_error(y_test, dt_predictions)
dt_r2 = r2_score(y_test, dt_predictions)

print("Decision Tree Regressor Results:")
print("Mean Squared Error:", dt_mse)
print("Mean Absolute Error:", dt_mae)
print("R-squared:", dt_r2)
```
